# 7. Cargar herramienta

```python
"""
                                    2D-GeoTemp 
Orlando Miguel Espinoza Ojeda1, Uriel Hazel Segura Gonzalez2, Erna Martha López Granados3 & Ruth Alfaro Cuevas Villanueva3

1CONAHCyT - Instituto de Investigaciones en Ciencias de la Tierra, Universidad Michoacana de San Nicolás de Hidalgo. (orlando.espinoza@umich.mx).
2Escuela Superior de Ingeniería y Arquitectura, Unidad Ticomán, Instituto Politécnico Nacional.(hazelsg99@gmail.com)
3Instituto de Investigaciones en Ciencias de la Tierra, Universidad Michoacana de San Nicolás de Hidalgo.


"""
import numpy as np 
import time
import matplotlib.pyplot as plt

print("Software para cálculo de perfiles de temperatura 2D")
print("Comienza la carga, espere")

def Perfil_2D_TEMPERATURA(xlong,ylong,Tnw,Tne,Tp,Grado_de_convergencia,Profundidades_de_Temp_W,Mediciones_de_Temp_W,Elevacion_w,fw,ge_w,Profundidades_de_Temp_E, Mediciones_de_Temp_E,Elevacion_e,fe,ge_e,Profundidades_de_Cond_W,Mediciones_de_Cond_W,kqw,Profundidades_de_Cond_E, Mediciones_de_Cond_E,kqe):
    
    # Elevación de ambos pozos para correlación de profunidades correcta
    if Elevacion_w == None:
        Elevacion_w=0
    if Elevacion_e == None:
        Elevacion_e=0
        
    dif_elevacion=np.rint(Elevacion_w)-np.rint(Elevacion_e)
    if dif_elevacion>0:
        Profundidades_de_Temp_E.insert(0, 0)
        Profundidades_de_Temp_E=dif_elevacion+Profundidades_de_Temp_E
        
        Mediciones_de_Temp_E.insert(0,Tne)
        Profundidades_de_Cond_E=dif_elevacion+Profundidades_de_Cond_E
        
        if Profundidades_de_Temp_E[-1]>ylong or Profundidades_de_Cond_E[-1]>ylong:
            ylong=max(Profundidades_de_Temp_E[-1],Profundidades_de_Cond_E[-1])
        
    elif dif_elevacion<0:
        Profundidades_de_Temp_W.insert(0, 0)
        Profundidades_de_Temp_W=dif_elevacion+Profundidades_de_Temp_W
        
        Mediciones_de_Temp_W.insert(0,Tnw)
        Profundidades_de_Cond_W=dif_elevacion+Profundidades_de_Cond_W 
        
        if Profundidades_de_Temp_W[-1]>ylong or Profundidades_de_Cond_W[-1]>ylong:
            ylong=max(Profundidades_de_Temp_W[-1],Profundidades_de_Cond_W[-1])
    
    
    
    nvcx=xlong
    nvcy=ylong
    # Crear matriz de Temperatura por Pozo
    Temp_W=np.column_stack((Profundidades_de_Temp_W, Mediciones_de_Temp_W))
    Temp_E=np.column_stack((Profundidades_de_Temp_E, Mediciones_de_Temp_E))
    
    # Crear matriz de Conductividad por Pozo
    Cond_W=np.column_stack((Profundidades_de_Cond_W, Mediciones_de_Cond_W))
    Cond_E=np.column_stack((Profundidades_de_Cond_E, Mediciones_de_Cond_E))
    
    # Conductividades
    con_pozo_w=Cond_W[:,1]
    con_pozo_e=Cond_E[:,1]
    
    dep_con_w=np.ones((len(Cond_W[:,0])+1,))
    dep_con_w[1:len(Cond_W[:,0])+1]=Cond_W[:,0]
    
    dep_con_e=np.ones((len(Cond_E[:,0])+1,))
    dep_con_e[1:len(Cond_E[:,0])+1]=Cond_E[:,0]
    
    # Temperatura
    val_fijos_w=Temp_W[:,1]
    val_fijos_e=Temp_E[:,1]
    
    dep_vel_w=np.ones((len(Temp_W[:,0])+1,))
    dep_vel_w[1:len(Temp_W[:,0])+1]=Temp_W[:,0]
    
    dep_vel_e=np.ones((len(Temp_E[:,0])+1,))
    dep_vel_e[1:len(Temp_E[:,0])+1]=Temp_E[:,0]
    
    num_cond_W=len(con_pozo_w)
    num_cond_E=len(con_pozo_e)
    num_val_w=len(val_fijos_w)
    num_val_e=len(val_fijos_e)
    
    # Cambio de variables
    Tn=Tnw
    Ti_fs=Tp
    xmaxdeltate_=Grado_de_convergencia
    kq=kqw
    
    if Cond_W[0,0]==None:
        CASE_C_W="AVERAGE"
    else:
        CASE_C_W="DEPTH"
        kq=sum(con_pozo_w)/len(con_pozo_w)
    
    if Cond_E[0,0]==None:
        CASE_C_E="AVERAGE"
    else:
        CASE_C_E="DEPTH"
        kqe=sum(con_pozo_e)/len(con_pozo_e)
    
    
    if Temp_W[0,0]==None:
        CASE_T_W="AVERAGE"
    else:
        CASE_T_W="DEPTH"
    
    if Temp_E[0,0]==None:
        CASE_T_E="AVERAGE"
    else:
        CASE_T_E="DEPTH"
    
    deltax=xlong/nvcx
    deltay=ylong/nvcy
    n=nvcx
    m=nvcy
    
    escala=1
    celdas=nvcx*nvcy
    num_de_escalados=[1]
    
    xlong_= xlong
    ylong_= ylong
    nvcx_= nvcx
    nvcy_= nvcy

    while celdas > 50000:
        escala=escala*10
        xlong_= int(xlong/escala)
        ylong_= int(ylong/escala)
        nvcx_= int(nvcx/escala)
        nvcy_= int(nvcy/escala)
        celdas=nvcx_*nvcy_
        num_de_escalados.append(escala)

    escala_sobrante=1
    etapa=2
    
    while escala!=0:
    
        print("Proceso en etapa {}".format(etapa), "de {}".format(len(num_de_escalados)+2))
        print("")
    
        inicio_g = time.time() 
    
        # PROFUNDIADES RESPECTIVAS PARA CONDUCTIVIDADES PARA POZO DE LA FRONTERA OESTE (W)
        if CASE_C_W=="DEPTH":
            dep_con_w_=np.rint((dep_con_w)/escala)
            dep_con_w_[0]=0
            if dep_con_w_[-1]==ylong_:
                dep_con_w_[-1]=dep_con_w_[-1]-1
    
        # PROFUNDIADES RESPECTIVAS PARA CONDUCTIVIDADES PARA POZO DE LA FRONTERA ESTE (E)
        if CASE_C_E=="DEPTH":
            dep_con_e_=np.rint((dep_con_e)/escala)
            dep_con_e_[0]=0
            if dep_con_e_[-1]==ylong_:
                dep_con_e_[-1]=dep_con_e_[-1]-1
                
        # PROFUNDIDAD RESPECTIVA DE VALORES FIJOS-CONDUCTIVOS DE TEMPERATURA DEL POZO      (W) 
        if CASE_T_W=="DEPTH":
            dep_vel_w_=np.rint((dep_vel_w)/escala)
            dep_vel_w_[0]=0
            if dep_vel_w_[-1]==ylong_:
                dep_vel_w_[-1]=dep_vel_w_[-1]-1
                
        # PROFUNDIDAD RESPECTIVA DE VALORES FIJOS-CONDUCTIVOS DE TEMPERATURA DEL POZO      (E) 
        if CASE_T_E=="DEPTH":
            dep_vel_e_=np.rint((dep_vel_e)/escala)
            dep_vel_e_[0]=0
            if dep_vel_e_[-1]==ylong_:
                dep_vel_e_[-1]=dep_vel_e_[-1]-1
                
        deltax_=xlong_/nvcx_
        deltay_=ylong_/nvcy_
    
        n_=nvcx_
        m_=nvcy_
    
        # CONDUCTIVIDADES PARA POZO DE LA FRONTERA OESTE (W) EN EL 2D: VARIABLES 
    
        k=np.ones((xlong_,ylong_))
    
        mit_n_=int(round(n_/2))
        if CASE_C_W=="DEPTH" :
            for l in range(0,num_cond_W):
                for i in range(0,mit_n_):
                    for j in range(int(dep_con_w_[l]),int(dep_con_w_[l+1])):
                        k[i,j]=con_pozo_w[l]
        elif CASE_C_W=="AVERAGE":
                k[0:mit_n_]=kq
    

        # CONDUCTIVIDADES PARA POZO DE LA FRONTERA ESTE (E) EN EL 2D:  VARIABLES
        if CASE_C_E=="DEPTH":
            for l in range(0,num_cond_E):
                for i in range(mit_n_,n_):
                    for j in range(int(dep_con_e_[l]),int(dep_con_e_[l+1])):
                        k[i,j]=con_pozo_e[l]
        elif CASE_C_E=="AVERAGE":
            k[mit_n_:n_]=kqe
         
        # **********************************************************************************************
        # 
        #       FRONTERA OESTE:  TW = Tc1   (DECLARACIÓN DE ARREGLOS PARA POZO OESTE)
        # 
        # **********************************************************************************************
    
        # CONDUCTIVIDADES TERMICAS 1D PARA EL POZO OESTE
        keac1=np.zeros((ylong_,))
    
        if CASE_C_W=="DEPTH":
                for l in range(0,num_cond_W):
                    for i in range(int(dep_con_w_[l]),int(dep_con_w_[l+1])):
                        keac1[i]=con_pozo_w[l]
        elif CASE_C_W=="AVERAGE":
            keac1[:]=kq
    
        # **********************************************************************************************
        # 	Calculos de flujo de calor para frontera Oeste, usando metodo TDMA
        #                     
        #                CALCULO DE FLUJO DE CALOR EN EL POZO OESTE W EN 1D UTILIZANDO 
        #                  ECUACIONES DE TRANSFERENCIA DE CALOR Y EL METODO DE TDMA
        # 
        # 
        # **********************************************************************************************
        #      PARAMETRO bw DE LA ECUACIÓN DE FLUJO DE CALOR
        # **********************************************************************************************
    
        if CASE_T_W=="DEPTH":
            if CASE_T_W=="AVERAGE":
                nvc_=nvcy_
                wa_=nvc_-1
            elif CASE_T_W=="DEPTH":
                nvc_=int(dep_vel_w_[-1])
                wa_=nvc_-1
            bw=np.ones((nvc_,))
            for i in range(0,wa_):
                bw[i]=keac1[i+1]
            bw[nvc_-1]=0
    
        # **********************************************************************************************
        #      PARAMETRO cw DE LA ECUACIÓN DE FLUJO DE CALOR
        # **********************************************************************************************
            cw=np.ones((nvc_,))
            cw[0]= 0
        # !	NODO (2, nvc-1)
            for i in range(1,nvc_):
                cw[i]=keac1[i-1]
    
        # **********************************************************************************************
        #      PARAMETRO aw DE LA ECUACIÓN DE FLUJO DE CALOR
        # **********************************************************************************************
            aw=np.ones((nvc_,))
        # NODO 1
            aw[0]= (((keac1[1])) + (2*(keac1[0])))
            
        # NODO (2, nvc-1)
            for i in range(1,wa_):
                aw[i]=(((keac1[i+1])) + ((keac1[i-1])))
    
        # NODO (nvc)
            aw[nvc_-1]=((2*(keac1[nvc_-1])) + ((keac1[nvc_-2])))
    
        # **********************************************************************************************
        #      PARAMETRO dw DE LA ECUACIÓN DE FLUJO DE CALOR
        # **********************************************************************************************
            dw=np.ones((nvc_,))
    
        # NODO 1
            dw[0]=(Tn*(2*(keac1[0])))
    
        # NODO (2, nvc-1)
            for i in range(1,wa_):
                dw[i]=0
    
        # NODO (nvc)
            dw[nvc_-1]=(val_fijos_w[-1]  *(2*(keac1[nvc_-1])))
    
        # **********************************************************************************************
        #      PARAMETRO Pw DE LA ECUACIÓN DE FLUJO DE CALOR
        # **********************************************************************************************
            Pw=np.ones((nvc_,))
            Pw[0]=bw[1-1]/aw[1-1]
    
            for i in range(1,wa_):
                Pw[i]= bw[i]/(aw[i]-cw[i]*Pw[i-1])
    
            Pw[nvc_-1]=0
    
        # **********************************************************************************************
        #      PARAMETRO Qw DE LA ECUACIÓN DE FLUJO DE CALOR
        # **********************************************************************************************
    
            Qw=np.ones((nvc_,))
    
            Qw[0]=dw[1-1]/aw[1-1]
    
            for i in range(1,nvc_):
                Qw[i]= (dw[i] + (cw[i]*Qw[i-1]))/(aw[i]-cw[i]*Pw[i-1]) 		
    
        # **********************************************************************************************
        #      REALIZAR CALCULO DE TEMPERATURAS EXTRAPOLADAS EN EL POZO OESTE = FRONTERA "W" OESTE
        # **********************************************************************************************
            Tc1=np.ones((m_,))
            Tc1[nvc_-1]=Qw[nvc_-1]
                                
            for i in range(wa_-1,-1,-1):
    
        # VALORES FIJOS-CONDUCTIVOS DE TEMPERATURA DEL POZO
                if CASE_T_W=="DEPTH":
                    for l in range(0,num_val_w):
                        Tc1[int(dep_vel_w_[l+1])]=val_fijos_w[l]
    
                Tc1[i]=(Pw[i]*Tc1[i+1]) + Qw[i]
    
        # Calculo de Resistividad Termica 
            if m_ > wa_-1 and escala==max(num_de_escalados):      
                Res_Termica_w=np.zeros((len(dep_vel_w_),))
    
                for i in range(0,len(val_fijos_w)):
                    if i ==0:
                        Res_Termica_w[i+1]=1/keac1[int(dep_vel_w_[i])]
                    else:
                        Res_Termica_w[i+1]=Res_Termica_w[i]+(1/keac1[int(dep_vel_w_[i])])
    
                Res_Termica_w[0]=0
                val_fijos_w_regrecion=[Tn]
                for i in range(0,len(val_fijos_w)):
                    val_fijos_w_regrecion.append(val_fijos_w[i])
    
                num_=len(Res_Termica_w)
                sum_x_y=sum(val_fijos_w_regrecion* Res_Termica_w)
                sum_x=sum(Res_Termica_w)
                sum_y=sum(val_fijos_w_regrecion)
                sum_x2=sum(Res_Termica_w**2)
    
                fw=  ((num_ * sum_x_y) - (sum_x * sum_y)) /(( num_ * sum_x2) - sum_x**2)
    
                plt.plot(Res_Termica_w,val_fijos_w_regrecion,"bo",label="Tem VS Res")
                plt.plot(Res_Termica_w,(fw*Res_Termica_w)+Tn,"r-",label="Flujo de calor = {}".format(round(fw, 2)))
                plt.grid()
                plt.legend(loc=4)
                plt.xlabel("Resistencia Termica")
                plt.ylabel("Temperatura")
                plt.xlim(left=0)
                plt.ylim(bottom=0)
                plt.title("POZO W - Res VS Tem")
                plt.show()
               
            fw_=(fw/1000)*escala 
            
            for i in range(0,m_):
                if i <= wa_-1:
                    Tc1[i]=(Pw[i]*Tc1[i+1]) + Qw[i]
                elif i > wa_-1:
                    Tc1[i] =  (fw_*1/kq) + Tc1[i-1]
                    
            if escala==max(num_de_escalados) and m_ > wa_-1:
    
                plt.plot(Tc1[0:wa_],(np.arange(0,-wa_*escala,-escala)),"b",label="Temperatura extrapolada en profundidad")
                plt.plot(Tc1[wa_:len(Tc1)],(np.arange(-wa_*escala,-m_*escala,-escala)),"r",label="Temperatura predecida")
                plt.grid()
                plt.legend(loc=3)
                plt.xlabel("Temperatura")
                plt.ylabel("Profundidad")
                plt.ylim(top=0)
                plt.title("Comportamiento de temperatura en Pozo W")
                plt.show()
    
        elif CASE_T_W=="AVERAGE":
            Tc1=np.ones((m_,))
            Tc1[0]=Tn
    
            if fw==None:    
                # CALCULO POR GRADIENTE GEOTERMICO
                gradi_geo_w=(ge_w/1000)*escala
    
                for i in range(1,m_):
                    Tc1[i]=Tc1[i-1]+gradi_geo_w
            else:
                # CALCULO POR FLUJO DE CALOR
                fw_=fw * escala
                
                for i in range(1,m_):
                    Tc1[i] =  (fw_*1/kq) + Tc1[i-1]
    
        if escala==max(num_de_escalados):
    
            plt.plot(Tc1,(range(0,-m_*escala,-escala)),"b",label="Temperatura")
            plt.grid()
            plt.legend(loc=3)
            plt.xlabel("Temperatura")
            plt.ylabel("Profundidad")
            plt.ylim(top=0)
            plt.title("Comportamiento de temperatura en Pozo W")
            plt.show() 
    
        # **********************************************************************************************
        # 	CCalculo de flujo de calor para prontera este, usando metodo TDMA
        #                     
        #                CALCULO DE FLUJO DE CALOR EN LA FRONTERA ESTE  TE = Tc3 EN 1D UTILIZANDO 
        #                  ECUACIONES DE TRANSFERENCIA DE CALOR Y EL METODO DE TDMA
        # 
        # 
        # **********************************************************************************************
        # 
        # 	ENTRADA DE DATOS
        # 
        # **********************************************************************************************
    
        # CONDUCTIVIDADES TÈRMICAS DEL POZO ESTE EN 1D
        if CASE_T_E=="DEPTH":
            keac2=np.ones((ylong_,))
            nvcy_=ylong_
    
            if CASE_C_E=="DEPTH":
                for l in range(0,num_cond_E):
                    for i in range(int(dep_con_e_[l]),int(dep_con_e_[l+1])):
                        keac2[i]=con_pozo_e[l]
            elif CASE_C_E=="AVERAGE":
                keac1[:]=kqe
    
            # **********************************************************************************************
            # 	PARAMETRO  be  DE LA ECUACION DE FLUJO DE CALOR
            # **********************************************************************************************
    
            if CASE_T_E=="AVERAGE":
                    nvce_=ylong_
            elif CASE_T_E=="DEPTH":
                    nvce_=int(dep_vel_e_[-1])
                    
            be=np.ones((nvce_,))
            wae_=nvce_-1
            # NODO (2, nvc-1)
            for i in range(0,wae_):
                be[i]=keac2[i+1]
    
            # NODO (nvc)
            be[nvce_-1]=0
    
            #    *********************************************************************************************
            # PARAMETRO  ce  DE LA ECUACION DE FLUJO DE CALOR
            #    *********************************************************************************************
    
            ce=np.ones((nvce_,))
            # NODO 1
            ce[0] = 0
    
            # NODO (2, nvc-1)
            for i in range(1,nvce_):
                ce[i]=(keac2[i-1])
    
            # **********************************************************************************************
            # PARAMETRO ae DE LA ECUACION DE FLUJO DE CALOR
            # **********************************************************************************************
            ae=np.ones((nvce_,))
            # NODO 1
            ae[0] = (keac2[2-1] + (2*keac2[1-1]))
    
            # NODO (2, nvc-1)
            for i in range(1,wae_):
                ae[i]=keac2[i+1] + keac2[i-1]
    
            # NODO (nvc)
            ae[nvce_-1]=2*keac2[nvce_-1] + keac2[nvce_-2]
    
            # **********************************************************************************************
            # !	PARAMETRO de DE LA ECUACION DE FLUJO DE CALOR
            # **********************************************************************************************
    
            de=np.ones((nvce_,))
            de[0]=Tne*2*keac2[1-1]
    
            # NODO (2, nvc-1)
            for i in range(1,wae_):
                de[i]= 0
    
            # NODO (nvc)
            de[nvce_-1]=(val_fijos_e[-1]*(2*(keac2[nvce_-1])))
    
            # **********************************************************************************************
            # PARAMETRO  Pe DE LA ECUACION DE FLUJO DE CALOR
            # **********************************************************************************************
            Pe=np.ones((nvce_,))
    
            Pe[0]=be[1-1]/ae[1-1]
    
            for i in range(1,wae_):
                Pe[i]= be[i]/(ae[i]-ce[i]*Pe[i-1])
    
            Pe[nvce_-1] = 0
    
            # **********************************************************************************************
            # PARAMETRO Qe DE LA ECUACION DE FLUJO DE CALOR
            # **********************************************************************************************
    
            Qe=np.ones((nvce_,))
    
            Qe[0]=de[1-1]/ae[1-1]
    
            for i in range(1,nvce_):
                Qe[i]= (de[i] + ce[i]*Qe[i-1])/(ae[i]-ce[i]*Pe[i-1])
    
            # **********************************************************************************************
            # CALCULAR TEMPERATURAS EXTRAPOLADAS EN EL POZO ESTE  FRONTERA TE = Tc3
            # **********************************************************************************************
    
            Tc3=np.ones((m_,))
            Tc3[nvce_-1]=Qe[nvce_-1]
    
            for i in range(wae_-1,-1,-1):
                # VALORES FIJOS-CONDUCTIVOS DE TEMPERATURA DEL POZO
                for l in range(0,num_val_e):
                    Tc3[int(dep_vel_e_[l+1])]=val_fijos_e[l]
    
                Tc3[i]=(Pe[i]*Tc3[i+1]) + Qe[i]
    
                # Calculo de Resistividad Termica 
            if m_ > wae_-1 and escala==max(num_de_escalados):      
                Res_Termica_e=np.zeros((len(dep_vel_e_),))
    
                for i in range(0,len(val_fijos_e)):
                    if i ==0:
                        Res_Termica_e[i+1]=1/keac2[int(dep_vel_e_[i])]
                    else:
                        Res_Termica_e[i+1]=Res_Termica_e[i]+(1/keac2[int(dep_vel_e_[i])])
    
                Res_Termica_e[0]=0
                val_fijos_e_regrecion=[Tne]
                for i in range(0,len(val_fijos_e)):
                    val_fijos_e_regrecion.append(val_fijos_e[i])
    
                num_=len(Res_Termica_e)
                sum_x_y=sum(val_fijos_e_regrecion* Res_Termica_e)
                sum_x=sum(Res_Termica_e)
                sum_y=sum(val_fijos_e_regrecion)
                sum_x2=sum(Res_Termica_e**2)
    
                fe=  ((num_ * sum_x_y) - (sum_x * sum_y)) /(( num_ * sum_x2) - sum_x**2)
    
                plt.plot(Res_Termica_e,val_fijos_e_regrecion,"go",label="Tem VS Res")
                plt.plot(Res_Termica_e,(fe*Res_Termica_e)+Tn,"r-",label="Flujo de calor = {}".format(round(fe,2)))
                plt.grid()
                plt.legend(loc=4)
                plt.xlabel("Resistencia Termica")
                plt.ylabel("Temperatura")
                plt.xlim(left=0)
                plt.ylim(bottom=0)
                plt.title("POZO E - Res VS Tem")
                plt.show()
              
            fe_=(fe/1000)*escala
    
            for i in range(0,m_):
                if i <= wae_-1:	
                    Tc3[i]=(Pe[i]*Tc3[i+1]) + Qe[i] 	 
                elif i > wae_-1:
                    Tc3[i] = (fe_*1/kqe) + Tc3[i-1]
    
            if escala==max(num_de_escalados) and m_ > wae_-1:
    
                plt.plot(Tc3[0:wae_],(range(0,-wae_*escala,-escala)),"g",label="Temperatura extrapolada en profundidad")
                plt.plot(Tc1[wae_:len(Tc3)],(range(-wae_*escala,-m_*escala,-escala)),"r",label="Temperatura predecida")
                plt.grid()
                plt.legend(loc=3)
                plt.xlabel("Temperatura")
                plt.ylabel("Profundidad")
                plt.ylim(top=0)
                plt.title("Comportamiento de temperatura en Pozo E")
                plt.show() 
               
        elif CASE_T_E=="AVERAGE":
            Tc3=np.ones((m_,))
            Tc3[0]=Tne
    
            if fe==None: 
                #CALCULO CON GRADIENTE GEOTERMICO
                gradi_geo_e=(ge_e/1000)*escala
    
                for i in range(1,m_):
                    Tc3[i]=Tc3[i-1]+gradi_geo_e  
            else:
                #CALCULO CON FLUJO DE CALOR            
                fe_=fe * escala
                
                for i in range(1,m_):
                    Tc3[i] =  (fe_*1/kq) + Tc3[i-1]
                
        if escala==max(num_de_escalados):
    
            plt.plot(Tc3,(range(0,-m_*escala,-escala)),"g",label="Temperatura")
            plt.grid()
            plt.legend(loc=3)
            plt.xlabel("Temperatura")
            plt.ylabel("Profundidad")
            plt.ylim(top=0)
            plt.title("Comportamiento de temperatura en Pozo E")
            plt.show() 
    
        # **********************************************************************************************
        # Calculo de flujo de calor para frontera sur
        #                     
        #                CALCULO DE FLUJO DE CALOR EN LA FRONTERA SUR  TS = Tc2 EN 1D UTILIZANDO 
        #                  ECUACIONES DE TRANSFERENCIA DE CALOR Y EL METODO DE TDMA
        # 
        # 
        # **********************************************************************************************
    
        # **********************************************************************************************
        # 	PARAMETRO  bs DE LA ECUACION DE FLUJO DE CALOR
        # **********************************************************************************************
    
        nvcs_=nvcx_
        bs=np.ones((nvcx_,))
        was_=nvcx_-1
    
        # NODO (2, nvc-1)
        k2=(kq + kqe )/2
        bs[:]=k2 
    
        # !	NODO (nvc)
        bs[nvcs_-1]=0
    
        # **********************************************************************************************
        # 	PARAMETRO  cs DE LA ECUACION DE FLUJO DE CALOR
        # **********************************************************************************************
        cs=np.ones((nvcs_,))
    
        # NODO (2, nvc-1)
        cs[:]=k2
    
        # NODO 1
        cs[0] = 0
    
        # **********************************************************************************************
        #  PARAMETRO as DE LA ECUACION DE FLUJO DE CALOR
        # **********************************************************************************************
        ass=np.ones((nvcs_,))
    
        # NODO 1
        ass[0] = k2 + 2*k2
    
        # NODO (2, nvc-1)
        for i in range(1,was_):
            ass[i]=k2 + k2
    
        # NODO (nvc)
        ass[nvcs_-1]=2*k2 + k2
    
        # **********************************************************************************************
        # 	PARAMETRO ds DE LA ECUACION DE FLUJO DE CALOR
        # **********************************************************************************************
        ds=np.ones((nvcs_,))
    
        # NODO 1   
        ds[0]=Tc1[-1]*2*k2
    
    
        # NODO (2, nvc-1)
        for i in range(1,was_):
            ds[i]= 0
    
        # NODO (nvc)
        ds[nvcs_-1]=Tc3[-1]*2*k2
        
        # **********************************************************************************************
        # PARAMETRO Ps DE LA ECUACION DE FLUJO DE CALOR
        # **********************************************************************************************
        Ps=np.ones((nvcs_,))
    
        Ps[0]=bs[1-1]/ass[1-1]
    
        for i in range(1,was_):
            Ps[i]= bs[i]/(ass[i]-cs[i]*Ps[i-1])
    
        Ps[nvcs_-1]= 0
        
        # **********************************************************************************************
        # 	PARAMETRO Qs DE LA ECUACION DE FLUJO DE CALOR
        # **********************************************************************************************
        Qs=np.ones((nvcs_,))
    
        Qs[0]=ds[1-1]/ass[1-1]
    
        for i in range(1,nvcs_):
            Qs[i] = (ds[i] + cs[i]*Qs[i-1])/(ass[i]-cs[i]*Ps[i-1])
    		
        # **********************************************************************************************
        #  CALCULO DE LA TEMPERATURA DE LA FRONTERA SUR
        # **********************************************************************************************
        Tc2=np.ones((nvcs_,))
    
        Tc2[nvcs_-1]=Qs[nvcs_-1]
    
        for i in range(was_-1,-1,-1):
            Tc2[i]=(Ps[i]*Tc2[i+1]) + Qs[i]
    
        # **********************************************************************************************
        # 	Calculo de temperatura 2D
        # 						CALCULO DE TEMPERATURAS EN EL PERFIL 2D 
        # **********************************************************************************************
    
        # DECLARACION DE VALORES DE TEMPERATURA DE LAS FRONTERAS DEL MODELO 2D
        twf=np.ones((nvcx_,nvcy_))
        tsf=np.ones((nvcx_,nvcy_))
        tef=np.ones((nvcx_,nvcy_))
        tnf=np.ones((nvcx_,nvcy_))
    
        # FRONTERA OESTE DEL 2D	
        for j in range(0,m_):
            for i in range(0,1):
                twf[i,j]=Tc1[j]
    
        #  FRONTERA SUR DEL 2D	   
        Tc4=np.ones((nvcs_,))
        gradiente_superficial=(Tn-Tne)/nvcx_
        Tc4[0]=Tn
        for i in range(1,nvcx_):
            Tc4[i]=Tc4[i-1]+gradiente_superficial
    
        for j in range(0,1):
            for i in range(0,n_):
                tsf[i,j]=Tc4[j]
        # FRONTERA ESTE DEL 2D	
        for j in range(0,m_):
            for i in range(n_-1,n_):
                tef[i,j]=Tc3[j]  
        # FRONTERA NORTE DEL 2D	
        for j in range(m_-1,m_):
            for i in range(0,n_):
                tnf[i,j]=Tc2[i]
    
        alfax=np.ones((nvcx_,nvcy_))
        alfay=np.ones((nvcx_,nvcy_))
        for i in range(0,n_):
            for j in range(0,m_):
                alfax[i,j]=k[i,j]*deltay_/deltax_
                alfay[i,j]=k[i,j]*deltax_/deltay_
    
        # **********************************************************************************************
        #      Temperaturas iniciales y de frontera supuesta
        # **********************************************************************************************
        tc=np.ones((n_,m_))
    
        for i in range(0,n_):
            for j in range(0,m_):    
                tc[i,j] =Ti_fs
                
        a=np.ones((n_,m_))
        b=np.ones((n_,m_))
        c=np.ones((n_,m_))
        d=np.ones((n_,m_))
        p=np.ones((n_,m_))
        q=np.ones((n_,m_))
        Tnew=np.ones((n_,m_))
    
        if escala!=max(num_de_escalados):
            tc=Tnew_escala.copy()
            
        deltatem=np.ones((n_,m_))
    
        a[0,0]=(3*alfax[0,0])+(3*alfay[0,0]) 
        a[0,m_-1]= (3*alfax[0,m_-1])+(3*alfay[0,m_-1]) 
        a[n_-1,0]= (3*alfax[n_-1,0])+(3*alfay[n_-1,0]) 
        a[n_-1,m_-1]= (3*alfax[n_-1,m_-1])+(3*alfay[n_-1,m_-1])
    
        for j in range(1,m_-1):
            a[0,j]=(3*alfax[0,j])+(2*alfay[0,j])
            a[n_-1,j]=(3*alfax[n_-1,j])+(2*alfay[n_-1,j])
    
        for i in range(1,n_-1):
            a[i,0]=(2*alfax[i,0])+(3*alfay[i,0])
            a[i,m_-1]=(2*alfax[i,m_-1])+(3*alfay[i,m_-1])	
    
        for j in range(1,m_-1):
            for i in range(1,n_-1):
                a[i,j]= 2.0*(alfax[i,j]+alfay[i,j])
    
        for j in range(0,m_):
            for i in range(0,n_-1):
                b[i,j]= alfax[i,j]
    
        for j in range(0,m_):
            b[n_-1,j]=0.0
    
        for j in range(0,m_):
            for i in range(1,n_):
                c[i,j]= alfax[i,j]
    
        for j in range(0,m_):
            c[0,j]=0.0
    
        # Calculo de la matriz tidiagonal TDMA 
    
        for j in range(0,m_):        
            p[0,j]= b[0,j]/a[0,j]
    
        for i in range(1,n_):
            for j in range(0,m_):
                p[i,j]= b[i,j]/(a[i,j]-(c[i,j]*p[i-1,j]))
    
        ciclo=1
        cont=0
        cont_estabilizado=0
        error=[n_]
        Error_comportamiento_error=[]
        error_permitido=n_/2
    
        #  ************************************************************************************
        #
        #  Primer procesado a escala, con el objetivo de tener un punto de partida cercano 
        #          al dato objetivo y de esta manera reducir tiempos de procesado. 
        #
        #  ************************************************************************************
        
        if escala==max(num_de_escalados):
            while ciclo!=0:
                
                # Caclulo de error para finalizar ciclo
                Error_comportamiento_error.append(max(error))
                if cont > 10:
                    prom_error=sum(Error_comportamiento_error[cont-10:cont])/10
                    max_error=max(Error_comportamiento_error[cont-10:cont])
                    min_error=min(Error_comportamiento_error[cont-10:cont])
    
                    if prom_error < error_permitido:
                        if max_error==prom_error and min_error==prom_error:
                            cont_estabilizado=cont_estabilizado+1
                            if cont_estabilizado > 3000:
                                ciclo=0
                        else:
                            cont_estabilizado=0
                
                #Calculo de perfil de TEMPERATURA
                d[0,0]=(2*alfax[0,0]*twf[0,0])+(2*alfay[0,0]*tsf[0,0])+(alfay[0,0]*tc[0,1])
                d[0,m_-1]=(2*alfax[0,m_-1]*twf[0,m_-1])+(2*alfay[0,m_-1]*tnf[0,m_-1])+(alfay[0,m_-1]*tc[0,m_-2])
                d[n_-1,0]=(2*alfax[n_-1,0]*tef[n_-1,0])+(2*alfay[n_-1,0]*tsf[n_-1,0])+(alfay[n_-1,0]*tc[n_-1,1])
                d[n_-1,m_-1]=(2*alfax[n_-1,m_-1]*tef[n_-1,m_-1])+(2*alfay[n_-1,m_-1]*tnf[n_-1,m_-1])+(alfay[n_-1,m_-1]*tc[n_-1,m_-2])
    
                for j in range(1,m_-1):
                    d[0,j]= (2*alfax[0,j]*twf[0,j])+(alfay[0,j]*tc[0,j+1])+(alfay[0,j]*tc[0,j-1])
                    d[n_-1,j]= (2*alfax[n_-1,j]*tef[n_-1,j])+(alfay[n_-1,j]*tc[n_-1,j+1])+(alfay[n_-1,j] *tc[n_-1,j-1])
    
                for i in range(1,n_-1):
                    d[i,0]= (2*alfay[i,0]*tsf[i,0])+(alfay[i,0]*tc[i,1])
                    d[i,m_-1]= (2*alfay[i,m_-1]*tnf[i,m_-1])+(alfay[i,m_-1]*tc[i,m_-2])
    
                for j in range(0+1,m_-1):
                    for i in range(1,n_-1):
                        d[i,j]= (alfay[i,j]*tc[i,j+1])+(alfay[i,j]*tc[i,j-1])
    
                for j in range(0,m_):
                    q[0,j]= d[0,j]/a[0,j]
    
                for i in range(1,n_):
                    for j in range(0+0,m_):
                        q[i,j]= (d[i,j]+(c[i,j]*q[i-1,j]))/(a[i,j]-(c[i,j]*p[i-1,j]))
    
                for j in range(0,m_):
                    Tnew[n_-1,j]= q[n_-1,j]
    
                for j in range(0,m_):
                    for i in range(n_-2,-1,-1):
                        Tnew[i,j]= (p[i,j]*Tnew[i+1,j])+q[i,j]    
    
                # **********************************************************************************************
                #  Criterio de convergencia
                for j in range(0+0,m_):
                    for i in range(0,n_):
                        deltatem[i,j] = abs(Tnew[i,j] - tc[i,j])  
            
                xmaxdeltate = xmaxdeltate_
    
                for j in range(0,m_):
                    for i in range(0,n_):
                        if deltatem[i,j] >  xmaxdeltate:
                            tc[i,j] = Tnew[i,j]
    
                comparison=tc[:,:] != Tnew[:,:]
                error=abs(sum(comparison) - n_)
                cont=cont+1
                
        #  ************************************************************************************
        #       
        #   Apartir del segundo procesado se ejecuta a hacer tantas iteraciones se puedan en 
        #                el mismo tiempo que demoro el primer procesado.
        #     Unicamente para adaptar y redistribuir el perfil de temperatura a escala superior
        #
        #  ************************************************************************************
        
        else:
            while ciclo!=0:
                if cont==0:               
                    inicio_ciclo_i= time.time()
                            
                d[0,0]=(2*alfax[0,0]*twf[0,0])+(2*alfay[0,0]*tsf[0,0])+(alfay[0,0]*tc[0,1])
                d[0,m_-1]=(2*alfax[0,m_-1]*twf[0,m_-1])+(2*alfay[0,m_-1]*tnf[0,m_-1])+(alfay[0,m_-1]*tc[0,m_-2])
                d[n_-1,0]=(2*alfax[n_-1,0]*tef[n_-1,0])+(2*alfay[n_-1,0]*tsf[n_-1,0])+(alfay[n_-1,0]*tc[n_-1,1])
                d[n_-1,m_-1]=(2*alfax[n_-1,m_-1]*tef[n_-1,m_-1])+(2*alfay[n_-1,m_-1]*tnf[n_-1,m_-1])+(alfay[n_-1,m_-1]*tc[n_-1,m_-2])
    
                for j in range(0+1,m_-1):
                    d[0,j]= (2*alfax[0,j]*twf[0,j])+(alfay[0,j]*tc[0,j+1])+(alfay[0,j]*tc[0,j-1])
                    d[n_-1,j]= (2*alfax[n_-1,j]*tef[n_-1,j])+(alfay[n_-1,j]*tc[n_-1,j+1])+(alfay[n_-1,j] *tc[n_-1,j-1])
    
                for i in range(1,n_-1):
                    d[i,0]= (2*alfay[i,0]*tsf[i,0])+(alfay[i,0]*tc[i,1])
                    d[i,m_-1]= (2*alfay[i,m_-1]*tnf[i,m_-1])+(alfay[i,m_-1]*tc[i,m_-2])
    
                for j in range(0+1,m_-1):
                    for i in range(1,n_-1):
                        d[i,j]= (alfay[i,j]*tc[i,j+1])+(alfay[i,j]*tc[i,j-1])
    
                for j in range(0+0,m_):
                    q[0,j]= d[0,j]/a[0,j]
    
                for i in range(1,n_):
                    for j in range(0,m_):
                        q[i,j]= (d[i,j]+(c[i,j]*q[i-1,j]))/(a[i,j]-(c[i,j]*p[i-1,j]))
    
                for j in range(0+0,m_):
                    Tnew[n_-1,j]= q[n_-1,j]
    
                for j in range(0+0,m_):
                    for i in range(n_-2,-1,-1):
                        Tnew[i,j]= (p[i,j]*Tnew[i+1,j])+q[i,j]    
    
                # **********************************************************************************************
                #     Criterio de convergencia
                for j in range(0+0,m_):
                    for i in range(0,n_):
                        deltatem[i,j] = abs(Tnew[i,j] - tc[i,j])    
    
                xmaxdeltate = xmaxdeltate_
                for j in range(0+0,m_):
                    for i in range(0,n_):
                        if deltatem[i,j] >  xmaxdeltate:
                            tc[i,j] = Tnew[i,j]
    
                comparison=tc[:,:] != Tnew[:,:]
                error=abs(sum(comparison) - n_)
                # Calculo de tiempo extra de procesado basado en tiempo a escala
                if cont==0:
                    fin__ciclo_i= time.time()
                    tiempo=(fin__ciclo_i-inicio_ciclo_i)
                    optimo=np.ceil(tiempo_en_ejecucion/tiempo)
    
                if cont >= optimo:
                    ciclo=0
                cont=cont+1
                
        # **********************************      
        # *****   Cambio de escala   *******
        # **********************************
    
        if (int(n_)*10*int(m_)*10) < 5000000:
            Tnew_escala=np.ones((n_*10,m_*10))
    
            for i in range(1,n_+1):
                for j in range(1,m_+1):
                    Tnew_escala[(i*10)-10:(i*10),(j*10)-10:(j*10)]=Tnew[i-1,j-1]
    
        final_e = time.time() 
        if escala==max(num_de_escalados):
            tiempo_en_ejecucion=(final_e-inicio_g)
    
            print("Tiempo aproximado a finalizar los cálculos de la herramienta es de {}".format(np.ceil(tiempo_en_ejecucion/60))," minutos")
            print("")
    
        if escala ==1:
            escala=0
        else:
            escala=escala/10
            xlong_= int(xlong/escala)
            ylong_= int(ylong/escala)
            nvcx_= int(nvcx/escala)
            nvcy_= int(nvcy/escala)
    
        celdas=nvcx_*nvcy_
        if celdas > 5000000:
            escala_sobrante=escala
            escala=0
            
            for i in range(0,(str(escala_sobrante).count('0'))):
                etapa=etapa+1
                print("Proceso en etapa {}".format(etapa), "de {}".format(len(num_de_escalados)+2))
                print("")
        etapa=etapa+1
        
    # ***********************************************************
    #
    # Ajuste de diferencia de elevacion entre pozos
    #
    # ***********************************************************
    Tnew_con_dif_elevacion=Tnew.copy()
    dimeciones_tnew=np.shape(Tnew)
    aumento_x_casilla=abs(dif_elevacion/dimeciones_tnew[0])
    num_nodos=0
    if dif_elevacion>0:
        for i in range(0,int(dimeciones_tnew[0])):
            if np.floor(num_nodos)==0:
                num_nodos+=aumento_x_casilla
            else:
                Tnew_con_dif_elevacion[i,0:int(np.floor(num_nodos))]=np.nan
                num_nodos+=aumento_x_casilla
                
    elif dif_elevacion<0: 
        for i in range(int(dimeciones_tnew[0])-1,-1,-1):
            if np.floor(num_nodos)==0:
                num_nodos+=aumento_x_casilla
            else:
                Tnew_con_dif_elevacion[i,0:int(np.floor(num_nodos))]=np.nan
                num_nodos+=aumento_x_casilla         
                          
        
    # **********************************************************************************************
    #
    #  Fin del procesado, se procede al ploteo y guardado de datos. 
    #
    # **********************************************************************************************
    
    print("Proceso en etapa {}".format(etapa), "de {}".format(len(num_de_escalados)+2))
    
    print("Etapa Final de guardado  de datos, la cual demorara dependiendo de la cantidad datos a guardar")
    print("En este caso el archivo de salida contendrá {} columnas".format(n_*m_))
    if escala_sobrante !=1:
        print("El tamañlo del mallado excede las posibilidades de la aplicación ONLINE, se recomienda para evitar este error descargar jupyter")
        print("Los archivos de salida se entregaran a escala 1:{}".format(int(escala_sobrante/10)))
    
    # Creacion de vector para el archivo .txt
    vector_tnew=np.zeros((np.size(Tnew),3))
    columna=0
    for i in range(0,m_):
        for j in range(0,n_):
            vector_tnew[columna,0]=j+1
            vector_tnew[columna,1]=i+1
            vector_tnew[columna,2]=Tnew[j,i]
            columna+=1
            
            
    # Guardado de archivo TXT
    np.savetxt(r'Matriz_de_Temperatura a escala.txt',vector_tnew, fmt='%.7g',delimiter="   ")
    
    # Ploteo y guardado de perfil de temperatura 2D
    if dif_elevacion==0:
        grafico_f=np.transpose(Tnew)
        color_map = plt.imshow(grafico_f)
        color_map.set_cmap("jet")
        plt.colorbar()
        plt.savefig("Grafico de Temperatura.png")  
        plt.show()
        
    if dif_elevacion!=0:
        grafico_f=np.transpose(Tnew)
        color_map = plt.imshow(grafico_f)
        color_map.set_cmap("jet")
        plt.colorbar()
        plt.title("Perfil de temperatura")
        plt.savefig("Grafico de Temperatura.png")  
        plt.show()
        
        EJEMPLO=np.transpose(Tnew_con_dif_elevacion)
        color_map = plt.imshow(EJEMPLO)
        color_map.set_cmap("jet")
        plt.title("Perfil de temperatura con dif de elevación")
        plt.colorbar()
        plt.savefig("Grafico de Temperatura con dif de elevación.png")  
        
    
    print("Proceso finalizado")
    
    print("")
    print("Se guardaron los siguientes documetos:")
    print("    Grafico de Temperatura.png")
    print("    Matriz de Temperatura.txt")

    return Tnew,Tnew_con_dif_elevacion
print("Software cargado, si en la celca posterior no existe error alguno, fue cargado correctamente")
```
